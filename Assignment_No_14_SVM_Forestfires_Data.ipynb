# Normalization function

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame
forestfires_norm = norm_func(forestfires)
forestfires_norm.tail()
```

# FEATURE ENGINEERING:-
### METHOD:- Feature Extraction with Recursive Feature Elimination.(RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
X = forestfires_norm.drop(["size_category"],axis = 1)
X.head()
```

```python
Y = forestfires_norm.iloc[:,-1]
Y 
```

```python
Feature_model = LogisticRegression()
fit = RFE(Feature_model,14).fit(X, Y)
```

```python
# Feature Ranking:
fit.ranking_
```

```python
X.head(2)
```

# Selecting Only Top 14 Features.

```python
X_feat = X.drop(["day","monthoct","monthsep"], axis=1)
X_feat.head()
```

```python
X1 = X_feat
Y1 = Y
```

# Building Model.
### 1) Grid Search CV

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_grid = {'C' : [0.1, 1, 10, 100, 1000, 10000], 'gamma' : [1, 0.1, 0.01, 0.001, 0.0001, 0.00001], 'kernel' : ['rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid,  cv = 10)
grid.fit(X1, Y1)
```

```python
print("Best Hyperparameter:-",grid.best_params_,"\n",
      "Best Accuracy Score:-",grid.best_score_)
```

#### Using This Best Hyperparameter For SVM Model Building Purpose. 

### 2) Building SVM Model.
#### a} Evaluate using Cross Validation.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kfold = KFold(n_splits = 30)
```

```python
Model = SVC( C = 10000, gamma = 0.01, kernel = "rbf")
results = cross_val_score(Model, X1, Y1, cv = kfold)
```

```python
print("Model Score by Cross Validation.:-",round(results.mean(),3)*100,"%")
```

#### b} Evaluate using Leave One Out Cross Validation.

```python
from sklearn.model_selection import LeaveOneOut
loocv = LeaveOneOut()
Model_SVM = SVC( C = 10000, gamma = 0.01, kernel = "rbf")
results_SVM = cross_val_score(Model_SVM, X, Y, cv = loocv)
results_SVM
```

```python
res1 = results_SVM.mean()*100.0
res1
```

```python
print("Model Score by Leave One Out Cross Validation :- ",np.round(res1,3),"%")
```
